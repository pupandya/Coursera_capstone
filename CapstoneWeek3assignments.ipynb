Getting data from the wiki page given

```python
import pandas as pd
```

```python
#specify the url
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
tables = pd.read_html(url)
```

Creating dataframe from the wiki data

```python
tables[0].head()
```

```python
df = tables[0]
df.head(10)
```

```python
df.columns
```

Giving appropriate columns

```python
df.columns = df.iloc[0]
```

```python
df.reindex(df.index.drop(0))
```

```python
df.describe()
```

Selecting the data not having Not assigned in Borough column

```python
df=df[df.Borough != 'Not assigned']
```

```python
df.describe()
```

```python
df.shape
```

Replace value of Borough to Neighbourhood where it is "Not assigned"

```python
df['Neighbourhood']=df['Neighbourhood'].replace('Not assigned', df['Borough'])
```
